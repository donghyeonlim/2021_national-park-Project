```python
import gpxpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('C:/Users/tripleler/Desktop/대곡2소공원3-천수봉_20210811_171337.gpx') as fh:
    gpx_file = gpxpy.parse(fh)
segment = gpx_file.tracks[0].segments[0]
coords = pd.DataFrame([
    {'lat': p.latitude,
     'lon': p.longitude,
     'ele': p.elevation,
     'time': p.time} for p in segment.points])
coords.head(3)
plt.plot(coords.lon[::1], coords.lat[::1], 'b-')
plt.show()
```

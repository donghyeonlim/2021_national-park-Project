```python
# 데이터 프레임 제작 / 시간 좀 걸림
df=pd.DataFrame(columns=(['file_name', 'max_conf', 'category', 'conf', 'bbox']))
for i in range(len(json_data['images'])):
    file_name=json_data['images'][i]['file'][44:][:-4]
    max_conf=json_data['images'][i]['max_detection_conf']
    if max_conf != 0: # 바운딩 박스가 그려진 사진이면
        category=json_data['images'][i]['detections'][0]['category']
        conf=json_data['images'][i]['detections'][0]['conf']
        bbox=json_data['images'][i]['detections'][0]['bbox']
    else: # 바운딩 박스가 그려지지 않은 사진이면
        category=np.nan
        conf=np.nan
        bbox=np.nan
    li=[file_name, max_conf, category, conf, bbox]
    df.loc[i]=li
df
```

```python
# 파일명으로 데이터프레임 오름차순 정렬
df=df.sort_values('file_name')
df
```

```python
# csv파일로 저장
df.to_csv('json.csv', header=True, index=False) # 칼럼명은 남기고 인덱스는 버림
```

```python
# 파일명에 '2마리' 단어가 포함된 데이터프레임 생성
# 바운딩 박스 두개도 잘 표현됨!
df2=pd.DataFrame(columns=(['file_name', 'max_conf', 'category', 'conf', 'bbox']))
for i in range(len(json_data['images'])):
    file_name=json_data['images'][i]['file'][44:][:-4]
    if '2마리' in file_name:
        max_conf=json_data['images'][i]['max_detection_conf']
        category=[]
        conf=[]
        bbox=[]
        if max_conf != 0:
            for j in range(len(json_data['images'][i]['detections'])):
                category.append(json_data['images'][i]['detections'][j]['category'])
                conf.append(json_data['images'][i]['detections'][j]['conf'])
                bbox.append(json_data['images'][i]['detections'][j]['bbox'])
        else:
            category=np.nan
            conf=np.nan
            bbox=np.nan
        li=[file_name, max_conf, category, conf, bbox]
        df2.loc[i]=li
df2
```

```python
# 전체 데이터프레임으로 다시 작성
df3=pd.DataFrame(columns=(['file_name', 'max_conf', 'category', 'conf', 'bbox']))
for i in range(len(json_data['images'])):
    file_name=json_data['images'][i]['file'][44:][:-4]
    max_conf=json_data['images'][i]['max_detection_conf']
    if max_conf != 0:
        if len(json_data['images'][i]['detections'])==1:
            category=json_data['images'][i]['detections'][0]['category']
            conf=json_data['images'][i]['detections'][0]['conf']
            bbox=json_data['images'][i]['detections'][0]['bbox']
        else:
            category=[]
            conf=[]
            bbox=[]
            for j in range(len(json_data['images'][i]['detections'])):
                category.append(json_data['images'][i]['detections'][j]['category'])
                conf.append(json_data['images'][i]['detections'][j]['conf'])
                bbox.append(json_data['images'][i]['detections'][j]['bbox'])
    else:
        category=np.nan
        conf=np.nan
        bbox=np.nan
    li=[file_name, max_conf, category, conf, bbox]
    df3.loc[i]=li
df3
```
